## Define Fitting Equations

```python
def L(x,A1,G,x0):
    return A1*(np.pi*G*(1+((x-x0)/G)**2))**-1;
def L2(x,X,Y,G1,G2,A,A1,A2):
    return L(x,A1,G1,X)+L(x,A2,G2,Y)+A;
def R(t,g,O,A,b):
    return A+np.exp(-g*t+b)*np.cos(O*t);
def T2_fit(t,A,T,a,B):
    return A*np.exp(-(t/T)**a)+B
```
